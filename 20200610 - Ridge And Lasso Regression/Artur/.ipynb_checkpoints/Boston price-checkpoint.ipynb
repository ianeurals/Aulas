# RIDGE REGRESSION

## Carregar Libs

```python
# dataframe and plot
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
# Carregar dataset
from sklearn.datasets import load_boston
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import r2_score, mean_squared_error
```

## Carregar dataset
* Basicamente o dataset trata sobre valores, entre outros atributos sobre casas no suburbio de Boston
### Detalhes sobre o dataset
    1. CRIM      per capita crime rate by town
    2. ZN        proportion of residential land zoned for lots over 25,000 sq.ft.
    3. INDUS     proportion of non-retail business acres per town
    4. CHAS      Charles River dummy variable (= 1 if tract bounds river; 0 otherwise)
    5. NOX       nitric oxides concentration (parts per 10 million)
    6. RM        average number of rooms per dwelling
    7. AGE       proportion of owner-occupied units built prior to 1940
    8. DIS       weighted distances to five Boston employment centres
    9. RAD       index of accessibility to radial highways
    10. TAX      full-value property-tax rate per 10,000;
    11. PTRATIO  pupil-teacher ratio by town
    12. B        1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
    13. LSTAT    % lower status of the population
    14. MEDV     Median value of owner-occupied homes in 1000's

```python
house_price = load_boston()
df = pd.DataFrame(house_price.data, columns=house_price.feature_names)
df['PRICE'] = house_price.target


df.head()
```

## Normalização e separação do dataset entre treino e teste
* A normalização feita ocorre devido o ridge penalizar seus coeficientes

```python
house_price.data = preprocessing.scale(house_price.data)
X_train, X_test, y_train, y_test = train_test_split(
    house_price.data, house_price.target, test_size=0.3, random_state=10)
```

### Gráfico com os dados normalizados

```python
# Quando o lambida tem o valor 0, então passa a ser o mesmo que o linear regression
ridge_reg = Ridge(alpha=0)
ridge_reg.fit(X_train, y_train)
ridge_df = pd.DataFrame({'variable': house_price.feature_names, 'estimate': ridge_reg.coef_})

ridge_train_pred = []
ridge_test_pred = []

# iteração
for alpha in np.arange(0, 200, 1):
    # treinamento
    ridge_reg = Ridge(alpha=alpha)
    ridge_reg.fit(X_train, y_train)
    var_name = 'estimate' + str(alpha)
    # Peso para cada feature
    ridge_df[var_name] = ridge_reg.coef_
    # predição
    ridge_train_pred.append(ridge_reg.predict(X_train))
    ridge_test_pred.append(ridge_reg.predict(X_test))

ridge_df = ridge_df.set_index('variable').T.rename_axis('estimate').rename_axis( 1).reset_index()
```

```python

fig, ax = plt.subplots(figsize=(15, 8))
ax.plot(ridge_df.RM, 'r', ridge_df.ZN, 'g', ridge_df.RAD, 'b', ridge_df.CRIM, 'c', ridge_df.TAX, 'y')
ax.axhline(y=0, color='black', linestyle='--')
ax.set_xlabel("Alpha")
ax.set_ylabel("Beta Estimate")
ax.set_title("Ridge Regression Trace")
ax.legend(labels=['Room','Residential Zone','Highway Access','Crime Rate','Tax'])
```

#### MSE
* MSE é um indicador para mensurar o erro. 
* Em linha verde temos o MSE voltado para o OLS e em vermelho para o Ridge

```python
ols_reg = LinearRegression()
ols_reg.fit(X_train, y_train)
ols_pred = ols_reg.predict(X_test)
ridge_mse_test = [mean_squared_error(y_test, p) for p in ridge_test_pred]
ols_mse = mean_squared_error(y_test, ols_pred)

plt.plot(ridge_mse_test[:25], 'ro')
plt.axhline(y=ols_mse, color='g', linestyle='--')
plt.title("Ridge Test Set MSE", fontsize=16)
plt.xlabel("Model Simplicity$\longrightarrow$")
plt.ylabel("MSE")
```

```python
lasso_reg = Lasso(alpha=0)
lasso_reg.fit(X_train, y_train)
lasso_df = pd.DataFrame({'variable': house_price.feature_names, 'estimate': ridge_reg.coef_})
print(lasso_df)

lasso_train_pred = []
lasso_test_pred = []
# iteração
for alpha in np.arange(0, 4, 1):
    # treinamento
    lasso_reg = Lasso(alpha=alpha)
    lasso_reg.fit(X_train, y_train)
    var_name = 'estimate' + str(alpha)
    # Peso para cada feature
    lasso_df[var_name] = lasso_reg.coef_
    # predição
    lasso_train_pred.append(lasso_reg.predict(X_train))
    lasso_test_pred.append(lasso_reg.predict(X_test))

lasso_df = lasso_df.set_index('variable').T.rename_axis('estimate').rename_axis(1).reset_index()
```

```python
fig, ax = plt.subplots(figsize=(15, 8))
ax.plot(lasso_df.RM, 'r', lasso_df.ZN, 'g', lasso_df.RAD, 'b', lasso_df.CRIM, 'c', lasso_df.TAX, 'y')
ax.axhline(y=0, color='black', linestyle='--')
ax.set_xlabel("Lambda")
ax.set_ylabel("Beta Estimate")
ax.set_title("Lasso Regression Trace")
ax.legend(labels=['Room','Residential Zone','Highway Access','Crime Rate','Tax'])
```

```python
ols_reg = LinearRegression()
ols_reg.fit(X_train, y_train)
ols_pred = ols_reg.predict(X_test)
lasso_mse_test = [mean_squared_error(y_test, p) for p in lasso_test_pred]
ols_mse = mean_squared_error(y_test, ols_pred)

plt.plot(lasso_mse_test[:25], 'ro')
plt.axhline(y=ols_mse, color='g', linestyle='--')
plt.title("Lasso Test Set MSE", fontsize=16)
plt.xlabel("Model Simplicity$\longrightarrow$")
plt.ylabel("MSE")
```

# EXTRA -- GRIDSEARCH CV

```python
from sklearn.model_selection import GridSearchCV
array_aux = []
for i in range(200):
    array_aux.append(i)
param_grid = {
    'alpha':[1,2,3,4,6,7,8,9,10,15,20,25,30,35,40,45,50,55,60,70,80,90,100,120,140,160,180,200],
    'normalize': [True, False],
   'fit_intercept':[True, False],
    'tol':[0.0001,0.001,0.02,0.1]
}
ridge = Ridge()
# cv = n-cross-validation
grid_search = GridSearchCV(estimator = ridge, param_grid = param_grid, 
                          cv = 3, n_jobs = -1, verbose = 2, refit = 'TRUE')
grid_search.fit(X_train, y_train)

print('Melhores parametros = \n',grid_search.best_params_)
print('\nMelhor pontuação: \n',grid_search.best_score_)
```
